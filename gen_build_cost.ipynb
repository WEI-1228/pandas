```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("gen_build_predetermined.csv")
```

```python
df.head(10)
```

```python
project_list = list(df["GENERATION_PROJECT"])
project_list[:10]
```

```python
single_project = set([x[:-5] for x in project_list])
single_project
```

```python
generation_2022 = [x + '-2022' for x in single_project]
df_2022 = pd.DataFrame(
    {
        "GENERATION_PROJECT": generation_2022,
        "build_year": [2022] * len(generation_2022),
        "gen_overnight_cost": [0] * len(generation_2022),
        "gen_fixed_om": [0] * len(generation_2022)
    }
)
df_2022.set_index("GENERATION_PROJECT", inplace=True)
df_2022["gen_overnight_cost"] = df_2022["gen_overnight_cost"].astype("float")
df_2022["gen_fixed_om"] = df_2022["gen_fixed_om"].astype("float")
```

```python
df_2022.head(10)
```

```python
data = {
    "Coal": [526267.805, 9999.088295],
    "Gas": [669180.897,	16060.341528],
    "Hydro_Nonpumped": [1133648.446, 17004.72669],
    "Wind": [993632.241, 14904.48362],
    "Central_PV": [709701.3, 10645.5195],
    "PV": [642662.46, 9639.9369],
    "Nuclear": [1954433.096, 39088.66192]
}
```

```python
for k, v in data.items():
    df_2022.loc[df_2022.index.str.contains(k), ("gen_overnight_cost", "gen_fixed_om")] = v
```

```python
df_2022.head(10)
```

```python
df_2023 = df_2022.copy()
df_2023.head()
```

```python
df_2023.index = [x + '-2023' for x in single_project]
df_2023.index.name = "GENERATION_PROJECT"
df_2023["build_year"] = ['2023'] * len(df_2023)
```

```python
df_2023.head(10)
```

```python
df_2022.sort_index(inplace=True)
df_2023.sort_index(inplace=True)
final_df = pd.concat((df_2022, df_2023))
```

```python
final_df.head(10)
```

```python
final_df.to_csv("gen_build_cost.csv")
```
