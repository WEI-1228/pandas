```python
import numpy as np
import pandas as pd
```

```python
name_dict = {
    "燃煤累计装机量": "Coal",
    "燃气累计装机": "Gas",
    "水电累计装机": "Hydro_Nonpumped",
    "风电累计装机": "Wind",
    "集中式光伏累计装机": "Central_PV",
    "分布式光伏累计装机": "PV",
    "核电累计装机": "Nuclear"
}
gen_can_retire_early_dict = {
    "燃煤累计装机量": 0,
    "燃气累计装机": 0,
    "水电累计装机": 1,
    "风电累计装机": 1,
    "集中式光伏累计装机": 1,
    "分布式光伏累计装机": 1,
    "核电累计装机": 0
}
```

```python
data_dict = pd.read_excel("新数据.xlsx", sheet_name=None)
for k, v in data_dict.items():
    data_dict[k] = v[["zone", "2022-mw"]].set_index('zone').sort_index().rename_axis("GENERATION_PROJECT")
    indexs = list(data_dict[k].index)
    new_indexs = {x:f"{x}-{name_dict[k]}-2022" for x in indexs}
    data_dict[k] = data_dict[k].rename(new_indexs)
    data_dict[k]["gen_can_retire_early"] = [gen_can_retire_early_dict[k]] * len(data_dict[k])
    data_dict[k]["build_year"] = [2022] * len(data_dict[k])
```

```python
data_dict["分布式光伏累计装机"]
```

```python
df1 = pd.concat(data_dict.values())
```

```python
df1 = df1.rename(columns={"2022-mw": "build_gen_predetermined"})
df1.head(10)
```

```python
df1 = df1.reindex(columns=["build_year", "build_gen_predetermined", "gen_can_retire_early"])
df1
```
